### Choosing a Client: Sync vs Async

The library provides two client options. **Both work identically except for the final `get_data()` call.**

| Client | Best For | When to Use |
|--------|----------|-------------|
| **`EcoPlots`** | Beginners, simple workflows | No `await` needed - simpler syntax |
| **`AsyncEcoPlots`** | Large datasets, production | Use `await` only for `get_data()` - faster |

**Key Point:** All methods (`select()`, `preview()`, `summary()`, etc.) work exactly the same in both clients. Only `get_data()` differs:
- `EcoPlots`: `df = ecoplots.get_data()`
- `AsyncEcoPlots`: `df = await ecoplots.get_data()`

````python
# ==================================================================
# Choose Your Client: Synchronous (simpler) or Asynchronous (faster)
# ==================================================================
# 
# **For beginners**: Use the synchronous `EcoPlots` client (uncomment below)
# - Simpler to use, no need for `await` keywords
# - Works immediately in notebooks and scripts
# - Best for small to medium datasets
#
# **For advanced users**: Use the asynchronous `AsyncEcoPlots` client (default)
# - Faster for large datasets and multiple requests
# - Requires `await` keyword ONLY for get_data()
# - Best for production workflows and large data processing
#
# **Important**: Both clients use identical code except for get_data()!
# - EcoPlots:      df = ecoplots.get_data()
# - AsyncEcoPlots: df = await ecoplots.get_data()
#
# Everything else (select, preview, summary, etc.) works exactly the same!
#
# To switch: Comment out one import and uncomment the other

# Option 1: Synchronous Client (Simpler - uncomment if preferred)
# from terndata.ecoplots import EcoPlots
# ecoplots = EcoPlots()

# Option 2: Asynchronous Client (Faster - currently active)
from terndata.ecoplots import AsyncEcoPlots
ecoplots = AsyncEcoPlots()

# Display the client information
print(ecoplots)
````

## 3. Applying Filters

The EcoPlots client uses a **fluent interface** for building queries. You can chain multiple filters together.

### 3.1 Dataset and Feature Type Selection

````python
# Select a dataset and feature type using fluent interface
ecoplots.select(dataset="TERN Ecosystem Surveillance")

# Display current configuration
print(ecoplots)
````

### 3.2 Feature Type Selection

````python
ecoplots.get_feature_types()
````

````python
ecoplots.select(feature_type="plant community")

print(ecoplots)
````

### 3.3 Region and Region Type Selection

````python
ecoplots.get_region_types()
````

````python
regions = ecoplots.get_regions("WWF ecoregions")
regions.head(10)
````

````python
# Apply region
ecoplots.select(region_type="WWF ecoregions", region="Pilbara shrublands")

print(ecoplots)
````

### 3.4 View Current Filters

Check what filters are currently active.

````python
# Get current filter configuration
filters = ecoplots.get_filter()
print("Active filters:")
for key, value in filters.items():
    print(f"  â€¢ {key}: {value}")

print("\n{}\n".format("="*79))

query_filters = ecoplots.get_api_query_filters()
print("Active API Query filters:")
for key, value in query_filters.items():
    print(f"  â€¢ {key}: {value}")
````

### 4.3 Get Full Dataset

Retrieve the complete dataset based on your filters.

````python
# ============================================================
# Retrieving Data: Sync vs Async
# ============================================================
#
# If using **synchronous** EcoPlots client:
#   df = ecoplots.get_data()
#
# If using **asynchronous** AsyncEcoPlots client (current):
#   df = await ecoplots.get_data()
#
# Note: `await` is only needed for AsyncEcoPlots!
# ============================================================

# Retrieve data asynchronously (returns GeoDataFrame)
df = await ecoplots.get_data()

print(f"Retrieved {len(df)} records")
print(f"Columns: {', '.join(df.columns)}")
df.head()
````

## 5. Advanced Features

### 5.1 Flexible Filter Selection

The EcoPlots library supports multiple ways to apply filters, giving you the flexibility to choose the approach that best fits your workflow. Whether you prefer precision with direct API URIs, readability with human-friendly names, or convenience with smart typo correction (with warnings), the library has you covered!

````python
# ============================================================
# Multiple Ways to Apply Filters
# ============================================================
# The library is flexible and supports various filtering approaches:
# 1. Direct API URIs (for precision)
# 2. Human-readable names (we'll resolve them)
# 3. Method chaining (for fluent queries)
# 4. Smart correction (common typos handled with warnings)
# ============================================================

# Clear previous selections to start fresh
ecoplots.clear()

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 1. Using Direct API URIs (Most Precise)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Use actual EcoPlots API endpoint URIs for exact filtering
ecoplots.select({
    "dataset": [
        "http://linked.data.gov.au/dataset/ausplots-forest"
    ],
    "project": [
        "http://linked.data.gov.au/dataset/ausplots-forest/dataset-fuel-load"
    ],
})

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 2. Method Chaining (Fluent Interface)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Chain multiple select() calls together for readable queries
ecoplots.select(site_id="NSFNNC0002").select(feature_type="plant litter")

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 3. Human-Readable Names + Smart Typo Correction
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Use natural names - the library will handle common typos
# Note: Corrections will display a warning showing what was fixed
ecoplots.select({
    "region_type": "NRM regions",
    "region": "North coasy",  # Typo: will be auto-corrected with a warning
})

# View the final configuration
print(ecoplots)
````
